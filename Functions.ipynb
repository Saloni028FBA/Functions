7. What are the advantages of using generators over regular functions?
  - The advantages of using generators over regular functions are as follows:

    1. Memory efficiency:
        
       - Regular function create and store data in memory before returning leading to excessive memory usage which is not suitable for large datasets or even memory loss, as its execution style is "immediate".
        
       - So the generator enter as hero in this case generating values only when demanded using "yeild" statement, the execution style is  "Lazy " which means generating and storing one value at a time, resulting in memory efficiency.
    
    2. Better Performance:
      
       - Since generator don't generate all value at once, they can be faster than the regular function as it take lot of time and making it slower than generators to give result as it first compute all the no and then providing result at once.

    3. Capability to represent Infinite squences:
      
       - Regular function cannot reprsent inifinte sequence as they would need to store an infinite number of values in memory even prone to memory loss.  
       
       - Generator are much more capable to represent infinite sequences because they produce value on demand saving huge storing spaces.

```python
# Example of regular function:(Ques-7)

def squares_list(n):
   for i in range(n):
    return i**2
```

```python
squares_list(10)
```

```python
# Example of generator function:(Ques-7)

def squares_gen(n):
    for i in range(n):
        yield i**2
```

```python
squares_gen(10)
```

```python
g = squares_gen(10)
```

```python
next(g)
```

```python
next(g)
```

```python
next(g)
```

8. What is a lambda function in Python and when is it typically used?
   - Lambda function is shorthand, anonymous function in python defined using in lambda keyword.
   - syntax: lambda args : expression.
   - Lambda function are concise, can be defined in one line.
   - They don't contain statement only expression, thus they can't have if or for loops inside lambda function.
   - Lambda function are particularly useful when single line or concise code is needed, especially working with high order functions like:
        1. map() - Map function execute a specified function for each of item of an iterable.
        2. filter() - Filter function is used to filter elements from iterable based on some conditions.

```python
# Example of Lambda function:(Ques-8)

nums = [1, 2, 3, 4, 5]    #map function
list(map(lambda x: x**2, nums))
```

```python
nums = [1, 2, 3, 4, 5]    #filter function
list(filter(lambda x : x % 2 == 0, nums))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
   - The map(), filter(), and reduce() functions in Python are powerful functional programming tools that operates on iterables, each serving a distinct purpose in data manipulation.
   - Here's a breakdown of their differences:
      
      1. Map function:
         
         - Applies a given function to each item in an iterable and returns an iterator that yields the results.
         - It generate returns transformed items or a map object (an iterator).
         - It's primarily used for transforming data within iterables.

      2. Filter function:
         
         -  The filter() function in Python is a built-in function that constructs an iterator from elements of an iterable for which a function returns true.
         - It is used to selects elements from an iterable based on a condition.
         - It generates returns a filter object (an iterator).
      
      3. Reduce function:
        
         - Applies a folding computation to sequential pairs of elements in an iterable, reducing them to a single value.
         - It generate returns one result or reducing an iterable to a single cumulative value.
         - It is use when you want to accumulate a result (like sum, product, etc.).  
         - Reduce function is not a built-in function, so should be exported from functools liabrary.

```python
# Example of map function:(Ques-10)

nums = [1, 2, 3, 4, 5]
list(map(lambda x: x**2, nums))
```

```python
# Example of filter function:(Ques-10)

nums = [1, 2, 3, 4, 5]
list(filter(lambda x: x % 2 != 0, nums))
```

```python
# Example of reduce function:(Ques-10)

from functools import reduce

nums = [1, 2, 3, 4]
(reduce(lambda x, y: x * y, nums))

```

```python
from functools import reduce

nums = [47, 11, 42, 13]
reduce(lambda x, y: x + y, nums)
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
# Ans 1.

def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
# Ans 6.
def powers_of_two(n):
    for exponent in range(n + 1):
        yield 2 ** exponent
```
